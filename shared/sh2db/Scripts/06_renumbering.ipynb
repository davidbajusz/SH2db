## Load in the pfam table:

```python
pfam = pd.read_csv(basedir+'data/pfam_table_recalculated_0830.csv')
pfam.drop(['Unnamed: 0'], 'columns', inplace = True)
#pfam.drop(['Unnamed: 0.1'], 'columns', inplace = True)
#pfam.drop(['Unnamed: 0.1.1'], 'columns', inplace = True)
pfam
```

## Renumber the pdb files: 

```python
for cat in pfam["Category"].unique():
    for gene in pfam[pfam["Category"] == cat]["Gene name"].unique():
        for pdb in pfam[pfam["Gene name"] == gene]["PDB ID"].unique():
            for chain in pfam[pfam["PDB ID"] == pdb]["PDB chain ID"].unique():
                for pole in pfam[(pfam["PDB ID"] == pdb) & (pfam["PDB chain ID"] == chain)]["Pole"].unique():
                    
                    os.system('cd /home/takacsg/Documents/sh2db/Structures/'+cat+'/'+gene+'/'+pdb+'/')
                    
                    pdb_id = str(pdb+'_'+chain+'_SH2'+pole)
                    print('PDB', pdb_id)
                    
                    increment1 = pfam[(pfam['PDB ID']==pdb) & (pfam["Pole"] == pole) & (pfam['PDB chain ID']==chain)]['New_uniprot_stop'].astype(int) - pfam[(pfam['PDB ID']==pdb) & (pfam["Pole"] == pole) & (pfam['PDB chain ID']==chain)]['New_PDB_stop'].astype(int)
                    
                    increment2 = pfam[(pfam['PDB ID']==pdb) & (pfam["Pole"] == pole) & (pfam['PDB chain ID']==chain)]['New_uniprot_start'].astype(int) - pfam[(pfam['PDB ID']==pdb) & (pfam["Pole"] == pole) & (pfam['PDB chain ID']==chain)]['New_PDB_start'].astype(int)
                    
                    increment = max(increment1.values[0],increment2.values[0])
                    print("INCREMENT ", increment)
                    increment = str(increment)
                    #print(type(increment))

                    ##shiftres = ('pdb_shiftres -'+increment+' /home/takacsg/Documents/sh2db/Structures/'+cat+'/'+gene+'/'+pdb+'/'+pdb+'_1_'+chain+'_SH2_'+pole+'.pdb > '+pdb+'_1_'+chain+'_SH2_'+pole+'_R.pdb')
                    shiftres = ('pdb_shiftres -'+increment+' /home/takacsg/Documents/sh2db/Structures/'+cat+'/'+gene+'/'+pdb+'/'+pdb+'_'+chain+'_SH2_'+pole+'.pdb > /home/takacsg/Documents/sh2db/Structures/'+cat+'/'+gene+'/'+pdb+'/'+pdb+'_'+chain+'_SH2_'+pole+'_R.pdb')
                                        
                    os.system(shiftres)
                    print('shiftres done for: ', pdb_id)
```
